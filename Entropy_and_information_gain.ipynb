```python
##imports

import pandas as pd
import numpy as np
```

```python
##data provided on the table

data = {
    "where": ["Home", "Home","Away", "Home", "Away","Away","Home","Home","Away","Home"],
    "when": ["7pm", "7pm", "7pm", "5pm", "9pm", "7pm","7pm", "7pm","7pm","9pm"],
    "Fred_Starts": ["Yes", "Yes", "Yes", "No", "Yes", "No","No","Yes","Yes","No"],
    "Joe_Offense": ["Center", "Forward","Forward", "Forward","Forward","Center", "Forward","Center","Center","Forward"],
    "Joe_Defense": ["Forward", "Center", "Forward", "Center", "Forward", "Forward", "Center","Center","Center","Center"],
    "Opp_C": ["Tall", "Short","Tall", "Tall", "Short","Tall","Tall","Tall","Short","Short"],
    "outcome": ["Won", "Won", "Won", "Lost", "Lost", "Won","Lost", "Won","Won", "Lost"],

}
##display data
pd.DataFrame(data)
```

```python
##Formula for calculating Entropy  = Probalility(log(2)*Probability)
##Formula for information gain = Outcome entropy - attributes entropy

##function for determining the best attribute
def best_attribute(outcome, attribute,attribute_name):

  ## calculating outcome entropy

  ##returns the probabilty of win and lost in the outcome column
  c = pd.Series(data['outcome']).value_counts(normalize = True)

  ##adds entropy(win + lost)
  outcome_entropy = np.sum(-(c * np.log2(c)))

  ##creates a dataframe of attribute and outcome
  data1 = pd.DataFrame(
      {
          "attribute": attribute,
          "outcome": outcome
      }
  )
  ##list the unique categories in the attributes selected
  values = list(data1['attribute'].unique())
  ##print("values are ", values)


  ##prints the length of the data in attributes
  total = len(data1['attribute'])
  ##print("values are ", total)

  attribute_entropy = 0

  for val in values:

    ##returns the total number of a given category of an attribute
    count = data1[data1['attribute'] == val]['outcome'].value_counts().sum()
    ##print(count)

    ##returns the probalities of win/lost of each individual category of an attribute
    probs = data1[data1['attribute'] == val]['outcome'].value_counts(normalize = True)
    ##print(probs)

    ##entropy of an attribute
    attribute_entropy += np.sum(-probs *np.log2(probs)) * (count/total)

  information_gain = outcome_entropy - attribute_entropy


  print("Attribute: {} \nEntropy: {:.4f} \nInformation Gain: {:.4f}\n\n\n".format(attribute_name, attribute_entropy, information_gain))

  return attribute_entropy, information_gain
```

```python
best_attribute(data['outcome'], data['when'], "when")
```

```python
## getting the best attribute
best_att = -np.inf #gets infinity value
attribute1 = ""

for key, value in data.items():
  if key == "outcome":
    continue

  ent, inf = best_attribute(data["outcome"], data[key], key)

  if inf > best_att:
    best_att = inf
    attribute1 = key

print(f"\n Best Attribute is {attribute1}  with information gain {best_att}")

```
